# Statistical Analysis
# Finding confidence interval for avg of total time spent on social media.

```python
social_media['Total Time Spent'].mean()
```

```python
samples =[]
stds = []
for i in range(30):
  try:
    x = social_media['Total Time Spent'].sample(10).values
    samples.append(x)
    stds.append(x)
  except Exception as e:
    print(e)
```

```python
samples = np.array(samples)
stds = np.array(stds)
```

```python
#applying CLT
sampling_dist_mean = samples.mean(axis = 1)
sampling_dist_stds = stds.std(axis = 1)
```

```python
sns.kdeplot(x = sampling_dist_mean)
```

```python
S = sampling_dist_stds.mean()
X = sampling_dist_mean.mean()
```

```python
#Applying t-procedure
#This will give the Confidence Interval
upper_lim = X - 2.042*(S/np.sqrt(30))
lower_lim = X + 2.042*(S/np.sqrt(30))
print('confidence interval : ' , upper_lim , " - " , lower_lim)
```

# CONCLUSION :- The Confidence Interval range from 123.124.. To 178.869..with 95% confidence level , which tells us that we're 95% sure that the average time spend on social media of people are most likely range in 123.124 To 178.869 mins a day.
